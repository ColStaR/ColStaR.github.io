## New Customer Metric

With the expansion location determined, we next need a measure of how many new customers are expected to be reached. In order to accomplish this, we used AGM's existing customer data from the Berkeley store in order to determine what percentage of people within the store's effective range were converted into AGM customers. 

As shown in the appendix section "Calculating New Customer Metric", the 15 mile radius of the Berkeley store reached a population of 2729857 and turned 7823 of them into customers. Thus, the new customer percentage was equal to 7823 / 2729857 = 0.29%. This indicates that for any population within a certain range, an average of 0.29% of them will be converted into AGM customers. As such, we will use that 0.29% of the total population in the region metric in order to conservatively estimate how many new customers are expected to be made with each delivery method. 

Do note that the new customer metric will differ for BART station pick-up, as BART stations have to account for two populations: the local population that lives within a range, and the ridership population that travels through the BART station. There was not a reliable source of data to indicate what the customer conversion rate would be for BART riders or for users living near BART stations, so we chose to modify the existing new customer conversion rate stated above.

In order to estimate the number of new customers gained from BART station ridership, we chose to use a new customer metric value of 0.5%. This value is higher than the regular new customer metric, as we believe that the close proximity and constant interaction with the AGM pick-up station in the BART station will serve as a form of advertisement and actually make customers more likely to purchase a meal compared to a standard delivery.

In order to estimate the number of new customers gained from the local population nearby a BART station, we chose to use a new customer metric value of 0.05%. This value is lower than the regular new customer metric because we believe the relative lack of proximity to the pick-up station for non-BART riders will make users less likely to purchase a meal compared to standard delivery. Furthermore, we intentionally made a very conservative estimate in order to avoid the possibility of double-counting new customers who are both BART riders and who live nearby.

### BART Station Pickup

Bay Area Rapid Transit is one of the most widely used public mass transit systems in the San Francisco Bay Area. With train stations all across the most populated regions around the bay, BART serves to shuttle hundreds of thousands of people as part of their daily commute. BART stations are distributed in cities across the bay, with many of the most busy stations being located in the most populated or active locations. Due to their distributed locality, connection to mass transit, and high volume of activity, BART Stations are logical choice to consider when expanding.

AGM would facilitate meal deliveries using a meal-safe locker system within or nearby a BART station that users would pick up meals from. This locker system will be refrigerated to ensure that the meal safety, automated to maximize availability, and possess a large number of lockers to ensure the maximum number of meals will be available. The Penguin Lockers COFRIDGE CF_24 model is being used as our example model for all calculations. This model of refrigerated locker meets all of the requirements and is currently available on the market.

For full notes and calculations regarding the BART Station Pickup kiosks, please refer to the appendix section "BART Station Data and Calculations" (Figure 4).

#### Effective Range

When determining the effective range for the BART Station Pickup locker, there are two groups of customers that need to be considered: the local population nearby that travels to the station to pick up meals, and the BART riders who interact with the locker as part of their BART ride.

It is difficult to determine the effective range of a locker kiosk on the local population, especially given that no similar data to use exists. However, we intuitively believe that the effective range for a locker would be 1 mile, as that is the longest distance that we believe an individual would be willing to walk to pick up their meal in the localized urban environment. As such, we are treating the effective range on the local population to be 1 mile.

BART riders are considered a captive audience, since their strict travel routes require users in a BART station to be nearby every element of the station. As such, we believe that the effective range of the lockers on BART riders is perfect: every BART rider that goes in or out of the station would be considered within the effective range.

#### Predicted New Customers

In order to determine how many new customers this would convert, we once again have to consider the two populations of the local population and the BART riders. Furthermore, we recognize that the new customer metric value may differ between the two populations, as the circumstances upon which both of them would interact with the locker are fundamentally different.

Within a BART station, we believe that the new customer metric would elevate above the standard 0.29% due to the unique circumstances of the BART riders. Given that the locker exists as a permanent and prominent fixture of the BART station that riders would see regularly, this will serve as a form of advertisement within that captive audience. As such, that will likely result in a greater level of curiosity, greater engagement, and thus a greater number of new customers for that population. As such, we estimate that the new customer metric would elevate to 0.5% of the total ridership population, thus equating to 212 new customers from the Montgomery Street station.

Within the local population nearby the BART station, we believe that the new customer metric would actually be lower than the standard percentage. Seeing as this local population does not ride BART, they will not be as immediately engaged with the locker. Furthermore, the lower footprint and lessened prominence of the locker compared to a physical location in a store also reduces visibility. However, this reduced visibility is more than made up by the sheer number of people nearby who will discovery and use the lockers eventually. As such, we estimate that the new customer metric would lower to 0.05% of the total local population, thus equating to 664 new customers from the local population around the Montgomery Street Station.

### Delivery Robot

Delivery robots are ground-based robots that move via tracks, wheels, or mechanical legs in order to deliver packages from one location to another. As industrialization and online ordering became highly prevalent, so too did the demand for package delivery. Delivery robots are touted as a viable alternative to traditional human or machinery-based delivery methods due to their low expense, accurate automation, and easy of scaling up. Companies such as Amazon, Boston Dynamics, and Kiwibot are spearheading the delivery robot industry.

AGM would facilitate meal deliveries with a delivery robot by having the robot deliver meals to customers from a centralized location. These delivery robots would need to be able to deliver meals to locations autonomously and rapidly, along with having the ability to control who can access the package to ensure accurate delivery. The Kiwibot 4.0 is one of the delivery robots that is having the most success, having been successfully trialed for use in Berkeley, CA. Because of this track record, Kiwibot 4.0 is an ideal example of a delivery robot to use.

For full notes and calculations regarding the delivery robots, please refer to the appendix section "Delivery Robot Data and Calculations" (Figure 5).

Delivery drones are touted as a viable alternative to traditional human or machinery-based delivery methods due to their low expense, accurate automation, easy of scaling up, and increased delivery speeds within certain ranges. # Limitation and Concerns

1. Our Population data is based on AGM's `zip_codes` table, and the data is not granular enough. The population is calculated based on the sum of the population of each zip code within X number of miles. This means it can include individuals in the zipcode but who live more than X miles from the station. A more granular data that provides a rough estimate for the number of people who lives within a given radius will provide more accurate results.  

2. AGM's customer location is based on their zip code saved in `customers` table, so we have the same limitation as Population data. A customer can live on the edge of a zip code and outside the range of X miles radius from the station but still gets included in our analysis. A more accurate customer location/ address will improve our analysis. Cell Phone GPS can capture the customer's Latitude and Longitude when the customer is registering and saved in DB.

3. The customer conversion percentage that was used to calculate the estimated number of new customer per delivery method was derived from the percentage of new customers from the Berkeley store. However, the number of new customers who use the Berkeley store may not be representative of the number of customers who will use the delivery methods. The Berkeley store's new customer conversion rate was used as a proxy for the lack of data regarding the number of new customers using the delivery methods; in the future, more research or better data would yield more accurate data.

4. The amount of estimated revenues per delivery method was calculated based on the sales data from the Berkeley store and applied to each of the delivery methods. However, the nature of the delivery methods is different from the Berkeley store, and thus the size and frequency of orders may differ between the two. No information was available regarding the size and frequency of delivery orders, so the Berkeley store data was used as a proxy; in the future, more research or better data would yield more accurate data.

# Technology and Trade-offs

During our research, we got a chance to use multiple technologies, and we want to discuss them briefly, show the trade-off we learned, and what's needed before we go to production and what would be best for future usage that fits the company growth. 

**SQL vs. NO-SQL** 

Our main DB is a SQL DB; that's where all of our data is stored. The data is stored in a 3rd normal form with no duplicates, which is great for faster writes since we only need to update the data in one place and also great for optimal storage since the data is stored once. 

At some point in our research, we needed all station data like Population around the station, Ridership, Yelp data, and the number of existing customers to be stored in one JSON formatted document. We needed that to provide faster analysis. So the usage of NO SQL DB would have been ideal in this case; we would sacrifice some redundancy for faster analysis. The workaround we did is to use python to build a JSON model that collects all the data for the SQL DB and other sources. 

With that in mind, we recommend setting up MongoDB Cluster. 

**Redis for Delivery tracking** 

Since we are introducing delivery, we might need to add Redis to our technology stack. 

During Drone or Robots delivery, we assume that the machine is equipped with GPS and would submit its location every few seconds. Later, the location gets viewed/ used by multiple other systems like the users themselves checking on the order location, our diagnostic system making sure that the machine is on its way and making progress, or the route mapping system to redirect the machine in case of new routes are available. 

With that in mind, a Redis cluster would be ideal for storing the location info. 

**Neo4j vs. SQL** 

Given that we are introducing delivery, we will need to calculate the shortest path from the pickup location to the customers to guarantee we are delivering in a timely fashion. 

That means we will have to start using Graphs. As much as SQL is helpful, it is not ideal for storing graphs, and doing any operation would be costly. 

With that in mind, a Neo4j cluster would be ideal for storing the graph data.
